### 백준 ( 2565 ) - 전깃줄
- ref : https://www.acmicpc.net/problem/2565

````python
N = int(input())
connect = {}
maxNum = 0
cnt = 0

for _ in range(N) : 
    a, b = map(int, input().split())
    connect[(a, b)] = abs(a-b)
    maxNum = max(a, b, maxNum)

sorted_dec = sorted(connect.items(), key = lambda item : item[1], reverse = True)

checkList_A = [0]*maxNum
checkList_B = [0]*maxNum

while len(sorted_dec) > 0 : 
    AB_flag = True
    (A, B), _ = sorted_dec.pop()
    
    print("* A : {}, B : {}".format(A, B))

    A, B = A-1, B-1

    if A > B : 
        A, B = B, A
        AB_flag = False

    start, end = (checkList_A, checkList_B) if AB_flag else (checkList_B, checkList_A)
    if AB_flag : 
        if 1 in end[A+1:B] : 
            print("POP!!!\n")
            cnt += 1
            continue
    else : 
        if 1 in start[A+1:B] : 
            print("POP!!!\n")
            cnt += 1
            continue
    
    start[A] = 1
    if A == B : 
        end[A] = 1
    else :
        end[B] = 1
        print("==========================")
    print("* index num   : {}".format([n for n in range(1, maxNum+1)]))
    print("--------------------------\n")
    print("* checkList_A : {}\n--------------------------".format(checkList_A))
    print("* checkList_B : {}\n==========================\n\n".format(checkList_B))
print(cnt if cnt < (N - cnt) else N-cnt)
````

## 최장 증가 부분 수열
### <p style="font-style:italic">Longest Increasing Subsequence ( ```LIS``` )</p>

> 원소가 <span style="color:green">n</span>개인 배열의 일부 원소를 골라내서 만든 부분 수열 중, 각 원소가 <span style="color:green">이전 원소보다 크다</span>는 조건을 만족하고, <span style="color:green">그 길이가 최대인 부분 수열</span>을 LIS라고 한다.

````python
# N개 입력 받기 => int형으로 저장
N = int(input())

# A, B 전선값을 저장하기 위한 list
record = []

# LIS를 계산하기 위해 N개 만큼 1로 채워진 list를 만든다 => [1, 1, 1, 1, ... ]
connect = [1 for _ in range(N)]

# 입력 받은 N개만큼 반복문을 돌면서 A, B을 입력받고 만들어둔 record에 추가를 해준다
for i in range(N) : 
    x, y = map(int, input().split())
    record.append([x, y])

# LIS를 풀기위해 저장된 [A, B] 형식 중에 A를 기준으로 정렬을 한다 
record.sort()

print("record : {}".format(record))

# LIS 찾기
# for(int i=0; i<N; i++)
for i in range(N) : 
    print('========================')
    print("i : {}".format(i))
    print("* before : {}".format(connect))
    
    # for(int j=i; j<i; j++)
    for j in range(i) : 
        print(f">> j : {j}")

        if record[i][1] > record[j][1] : 
            # ex) max( 1, 1+1 ) ==> 2
            connect[i] = max(connect[i], connect[j] + 1)
    print("* after  : {}".format(connect))

print(N-max(connect))
````
